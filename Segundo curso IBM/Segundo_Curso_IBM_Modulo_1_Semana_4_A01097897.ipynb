# Sección 1 - Notebooks

Se usará el dataset disponible en https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data

Procedemos a cargarlo y manipularlo usando python

```python
#El primer paso del laboratorio lo vamos a saltar ya que no necesitamos instalar las librerías de pandas o numpy - Google colab ya las tiene
#Importamos las librerías que vamos a usar
import pandas as pd
import numpy as np
```

```python
#Procedemos a descargar el archivo a nuestro ambiente en el google colab
#Este paso del laboratorio también lo saltaremos ya que google colab no permite la conexión usando pyodide.http
#En su lugar vamos a leer el archivo directamente desde el path

path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"

df = pd.read_csv(path, header=None)
```

```python
#Imprimimos las primeras 5 líneas del archivo para validar que se cargó correctamente
# show the first 5 rows using dataframe.head() method
print("The first 5 rows of the dataframe") 
df.head(5)
```

**Pregunta 1: Imprime los últimos 10 registros del archivo**

```python
df.tail(10)
```

```python
#El siguiente paso es agregar los encabezados de las columnas. Nos ayudará a comprender mejor el dataset

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
df.columns = headers
df.head(10)
```

```python
#Reemplazamos el question mark ? por NaN con la función de pandas replace
df1 = df.replace('?',np.NaN)
```

```python
#También podemos eliminar el registro completo si el valor de cierta columna para dicho registro es nulo
df = df1.dropna(subset=["price"], axis=0)
df.head(20)
```

**Pregunta 2: Encuentra el numbre de las columnas del dataframe**


```python
#La variable columns del objeto dataframe contiene la lista indexada de cada columna
df.columns
```

```python
#Se puede guardar el dataset en otro archivo de salida 
#Para este caso se guardará en otro archivo csv con la función to_csv
df.to_csv("automobile.csv", index=False)
```

**Aspectos básicos de un DataSet**


```python
#Podemos observar el data type de cada columna
df.dtypes
```

```python
#También podemos describir el dataset. Esto nos devolverá análisis estadístico de las columnas numéricas del dataset
df.describe()
```

```python
#Si incluimos el argumento indlude = all, también hará análisis en aquellas columnas no numéricas. 
#Desde luego, no será capaz de realizar análisis numéricos en columnas no numéricas

df.describe(include = 'all')
```
